```python
def Stem_text(text):

    # Tokenize the input text into words
    tokens = word_tokenize(text)
    stemmed_tokens = [stemmer.stem(word) for word in tokens]
    # print (tokens)
    return ' '.join(stemmed_tokens)

def clean(text):
  # Remove special characters, tabs, line jumps, and extra white spaces
   text = re.sub(r"[\.\,\#_\|\:\?\?\/\=\@]", " ", text) # remove special characters
   text = re.sub(r'\t', ' ', text) # remove tabs
   text = re.sub(r'\n', ' ', text) # remove line jump
   text = re.sub(r"\s+", " ", text) # remove extra white space
   text = text.strip()
   return text

def clean(text):
    text = re.sub(r"[\.\,\#_\|\:\?\?\/\=\@']", " ", text)  # Include "'" in the regex pattern
    text = re.sub(r'\t', ' ', text)
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r"\s+", " ", text)
    text = text.strip()
    return text

def remove_stopwords(text):
    tokens = word_tokenize(text)
    # Filter out the stopwords, normalizing by converting to lowercase
    filtered_tokens = [word.lower() for word in tokens if word.lower() not in stop_words]
    return ' '.join(filtered_tokens)

def preprocess(sentence):
  sentence = clean(sentence)
  sentence = remove_stopwords(sentence)
  sentence = Stem_text(sentence)
  return sentence
```

```python
def load_cisi_dataset(data_dir):
    documents_path = os.path.join(data_dir, 'CISI.ALL')
    queries_path = os.path.join(data_dir, 'CISI.QRY')
    qrels_path = os.path.join(data_dir, 'CISI.REL')

    documents_df = read_documents(documents_path)
    queries_df = read_queries(queries_path)
    qrels_df = read_qrels(qrels_path)
    return documents_df, queries_df, qrels_df

# Read documents from CISI.ALL file
def read_documents(documents_path):
    with open(documents_path, 'r') as file:
        lines = file.readlines()
    documents = []
    current_document = None
    for line in lines:
        if line.startswith('.I'):
            if current_document is not None:
                current_document['Text'] = current_document['Text'].split('\t')[0].strip()  # Remove anything after the first tab
                documents.append(current_document)
            current_document = {'ID': line.strip().split()[1], 'Text': ''}
        elif line.startswith('.T'):
            continue
        elif line.startswith('.A') or line.startswith('.B') or line.startswith('.W') or line.startswith('.X'):
            continue
        else:
            current_document['Text'] += line.strip() + ' '

    # Append the last document
    if current_document is not None:
        current_document['Text'] = current_document['Text'].split('\t')[0].strip()  # Remove anything after the first tab
        documents.append(current_document)
    documents_df = pd.DataFrame(documents)
    return documents_df

# Read queries from CISI.QRY file
def read_queries(queries_path):
    with open(queries_path, 'r') as file:
        lines = file.readlines()
    query_texts = []
    query_ids = []
    current_query_id = None
    current_query_text = []
    for line in lines:
        if line.startswith('.I'):
            if current_query_id is not None:
                query_texts.append(' '.join(current_query_text))
                current_query_text = []
            current_query_id = line.strip().split()[1]
            query_ids.append(current_query_id)
        elif line.startswith('.W'):
            continue
        elif line.startswith('.X'):
            break
        else:
            current_query_text.append(line.strip())
    # Append the last query
    query_texts.append(' '.join(current_query_text))
    queries_df = pd.DataFrame({
        'qid': query_ids,
        'raw_query': query_texts})
    return queries_df

# Read qrels from CISI.REL file
def read_qrels(qrels_path):
    qrels_df = pd.read_csv(qrels_path, sep='\s+', names=['qid','Q0','docno','label'])
    return qrels_df
```

```python
# Initialize a dictionary to store term frequencies

term_freq = {}

# Iterate over each row in the documents DataFrame
for index, row in documents_df.iterrows():

    # Split the processed text into individual terms
    terms = row['processed_text'].split()

    for term in terms:

        if term in term_freq:
            term_freq[term]['frequency'] += 1
            term_freq[term]['document_ids'].append(row['docno'])
        else:
            term_freq[term] = {'frequency': 1, 'document_ids': [row['docno']]}

for term, info in term_freq.items():
    print(f"Term: {term}")
    print(f"Frequency: {info['frequency']}")
    print(f"Document IDs: {info['document_ids']}")
```

```python
# Initialize the TF-IDF retrieval model

tfidf_retr = pt.BatchRetrieve(index_path, controls={"wmodel": "TF_IDF"}, num_results=10)
```

```python
from flask import Flask, request, render_template_string
import pandas as pd
from datetime import datetime
import torch

app = Flask(__name__)

# Define the preprocess function
def preprocess(query):
    sentence = clean(query)
    sentence = remove_stopwords(sentence)
    sentence = Stem_text(sentence)
    return sentence

# Define the HTML template
HTML = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Cisi Search Engine</title>
    <style>
        body { font-family: Arial, sans-serif; margin: 20px; }
        h1 { color: #333; }
        form { margin-bottom: 20px; }
        .result { margin-bottom: 20px; }
    </style>
</head>
<body>
    <h1>Type your query to reveal the data's documents</h1>

    <form action="/search" method="POST">
        <input type="text" name="query" placeholder="Enter your search query" required style="width: 300px; padding: 8px;">
        <button type="submit" style="padding: 8px;">Search</button>
    </form>

    {% if query %}
        <h2>Results for "{{ query }}":</h2>
        <p>About {{ efficiency }} results ({{ effectiveness }} seconds)</p>

        {% if results %}
            {% for doc_id, content in results.items() %}
                <div class="result">
                    <h3>Document ID: {{ doc_id }}</h3>
                    <p>Content: {{ content }}</p>
                </div>
            {% endfor %}
        {% else %}
            <p>No results found.</p>
        {% endif %}
    {% endif %}
</body>
</html>
"""

@app.route("/")
def home():
    return render_template_string(HTML)

@app.route("/search", methods=['POST'])
def search():
    if 'query' in request.form:
        query = request.form['query']
    else:
        raise ValueError("Query not found in form data.")

    start_time = datetime.now()

    # Preprocess the query
    query = preprocess(query)

    expanded_query = rm3_qe.search(query).iloc[0]["query"]
    expanded_query_formatted = ' '.join(expanded_query.split()[1:])

    results = tfidf_retr.search(expanded_query_formatted)

    # Merging results with documents_df
    result_merged = pd.merge(results, documents_df, on="docno", how="left")

    # Sort by TF-IDF score and take top 5 results for BERT reranking
    top_results = result_merged.sort_values(by="score", ascending=False).head(5)

    # Prepare the input for BERT
    inputs = tokenizer(top_results["processed_text"].tolist(), padding=True, truncation=True, return_tensors="pt", max_length=512)

    # Get scores from BERT
    with torch.no_grad():
        outputs = model(**inputs, return_dict=True)
    scores = outputs.logits[:, 1]  # Assuming binary classification and index 1 is the relevant class

    # Process BERT output and sort results
    top_results['bert_score'] = scores.detach().cpu().numpy()
    top_results = top_results.sort_values(by="bert_score", ascending=False)

    end_time = datetime.now()
    execution_time = (end_time - start_time).total_seconds()

    # Prepare results to pass to the template, including document content
    results_dict = {}
    for _, row in top_results.iterrows():
        results_dict[row['docno']] = row['processed_text']

    return render_template_string(HTML, query=query, results=results_dict, efficiency=len(results_dict), effectiveness=execution_time)

@app.route("/document/<doc_id>", methods=["GET"])
def doc_html(doc_id):
    text_content = documents_df[documents_df['docno'] == doc_id]['processed_text'].iloc[0]
    text_html = f"{text_content}"
    return render_template_string(f"""
    <h1>Document {doc_id}</h1>
    <p>{text_html}</p>
    """)

if __name__ == '__main__':
    app.run(port=5000)
```
